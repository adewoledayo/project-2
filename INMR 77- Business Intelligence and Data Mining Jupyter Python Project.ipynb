```python
# DATA PREPARATION & TRANSFORMATION


from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the 'object_column' to float
AirBnB_Data['host_is_superhost'] = label_encoder.fit_transform(AirBnB_Data['host_is_superhost']).astype(float)
AirBnB_Data['host_has_profile_pic'] = label_encoder.fit_transform(AirBnB_Data['host_has_profile_pic']).astype(float)
AirBnB_Data['host_identity_verified'] = label_encoder.fit_transform(AirBnB_Data['host_identity_verified']).astype(float)
AirBnB_Data['has_availability'] = label_encoder.fit_transform(AirBnB_Data['has_availability']).astype(float)
AirBnB_Data['instant_bookable'] = label_encoder.fit_transform(AirBnB_Data['instant_bookable']).astype(float)
AirBnB_Data['room_type'] = label_encoder.fit_transform(AirBnB_Data['room_type']).astype(float)

# Print the updated DataFrame
AirBnB_Data
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine
writer = pd.ExcelWriter('AirBnB_Data.xlsx', engine='xlsxwriter')

# Write the DataFrame to an Excel sheet named "Sheet1"
AirBnB_Data.to_excel(writer, sheet_name='AirBnB_Data')

# Save the Excel file and close the writer
writer.save()
```
